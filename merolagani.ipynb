```python
from bs4 import BeautifulSoup
import requests
```

```python
url ="https://merolagani.com/latestmarket.aspx"
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')


print(soup)
```

```python
# Find all tables on the page
world_titles = soup.find_all('table')
```

```python
table = soup.find('table', class_='table table-hover live-trading sortable')
# table = soup.find('table', class_='table table-hover table-index')
```

```python
world_titles = table.find_all('th')
```

```python
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles if title.find('span')]

print(world_table_titles)
```

```python
import pandas as pd
df = pd.DataFrame(columns = world_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```

```python

df
```

```python
df.to_csv(r'output.csv', index = False)
```
